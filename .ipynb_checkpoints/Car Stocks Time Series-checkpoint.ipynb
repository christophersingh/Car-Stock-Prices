```python
import pandas_datareader as pdr

def getData(symbol):
    start_date = '20190101'
    end_date = '20191231'

    timeDF = pdr.get_data_yahoo(symbol, start=start_date, end = end_date)
    timeDF.drop('Adj Close', axis=1, inplace=True)
    return timeDF
    
BMW = getData('BMWYY')
Audi = getData('AUDVF')
Porsche = getData('POAHY')
Mercedes = getData('DMLRY')
Volvo = getData('VLVLY')
Tesla = getData('TSLA')
Volkswagen = getData('VWAGY')
Nissan = getData('NSANY')
Fiat = getData('FCAU')
AstonMartin = getData('ARGGY')

BMW['StockName'] = 'BMW'
Audi['StockName'] = 'Audi'
Porsche['StockName'] = 'Porsche'
Mercedes['StockName'] = 'Mercedes'
Volvo['StockName'] = 'Volvo'
Tesla['StockName'] = 'Tesla'
Volkswagen['StockName'] = 'Volkswagen'
Nissan['StockName'] = 'Nissan'
Fiat['StockName'] = 'Fiat'
AstonMartin['StockName'] = 'Aston Martin'

cars = [BMW, Audi, Porsche, Mercedes, Volvo, Tesla, Volkswagen, Nissan, Fiat, AstonMartin]
```

```python
for x in cars:
    with plt.style.context('ggplot'):
        plt.figure(figsize=(8,6))
        plt.title(str(x['StockName'][0]))
        plt.plot(x.Close, label=x)    
```

```python
mva1 = 5
mva2 = 20

for x in cars:
    x['ma_short'] = x.Close.rolling(mva1, min_periods = 1).mean()
    x['ma_long'] = x.Close.rolling(mva2, min_periods = 1).mean()
    print(x)
```

```python
for data in cars:
    with plt.style.context('ggplot'):
        plt.figure(figsize=(18,16))
        plt.title(str(data['StockName'][0]))
        plt.plot(data.Close, label='Microsoft')
        plt.plot(data.ma_short, label='MA Short')
        plt.plot(data.ma_long, label='MA Long')
```

```python
import numpy as np
for data in cars:
    data.dropna(inplace = True)
    data['trade'] = np.where(data['ma_short']> data['ma_long'], 1, -1)
```

```python
for data in cars:
    data['return'] = np.log(data.Close).diff()
```
